# Top 10 Countries: Radar/Spider Plot

```python
from math import pi

# Prepare data
top10 = rich_country_df.sort_values(by='gdp_per_capita', ascending=False).head(10)
categories = ['gdp_per_capita', 'unemployment_rate', 'cost_index', 'purchasing_power_index']
radar_data = []
for idx, row in top10.iterrows():
    country = row['country']
    values = [
        row['gdp_per_capita'],
        unemployment_df[unemployment_df['country'] == country]['unemployment_rate'].values[0] if country in unemployment_df['country'].values else 0,
        cost_of_living_df[cost_of_living_df['country'] == country]['cost_index'].values[0] if country in cost_of_living_df['country'].values else 0,
        cost_of_living_df[cost_of_living_df['country'] == country]['purchasing_power_index'].values[0] if country in cost_of_living_df['country'].values else 0
    ]
    radar_data.append((country, values))

# Normalize values for radar plot
import numpy as np
values_array = np.array([v for _, v in radar_data])
values_norm = (values_array - values_array.min(axis=0)) / (values_array.max(axis=0) - values_array.min(axis=0) + 1e-9)

# Radar plot
plt.figure(figsize=(8,8))
angles = [n / float(len(categories)) * 2 * pi for n in range(len(categories))]
angles += angles[:1]
for i, (country, values) in enumerate(zip([c for c, _ in radar_data], values_norm)):
    vals = list(values) + [values[0]]
    plt.polar(angles, vals, label=country)
plt.xticks(angles[:-1], categories)
plt.title("Radar Plot: Top 10 Countries by GDP per Capita")
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
plt.show()
```

```python
import squarify

# If you have a 'region' column, otherwise skip or add it
if 'region' in rich_country_df.columns:
    region_gdp = rich_country_df.groupby('region')['gdp_per_capita'].sum().reset_index()
    plt.figure(figsize=(12,6))
    squarify.plot(sizes=region_gdp['gdp_per_capita'], label=region_gdp['region'], alpha=.8)
    plt.title("Treemap of GDP per Capita by Region")
    plt.axis('off')
    plt.show()
```

# Bubble Plot: GDP vs. Cost Index vs. Unemployment

```python
plt.figure(figsize=(12,8))
plt.scatter(
    merged_df['gdp_per_capita'],
    merged_df['cost_index'],
    s=merged_df['unemployment_rate']*20,  # Bubble size
    alpha=0.6
)
plt.xlabel('GDP per Capita')
plt.ylabel('Cost Index')
plt.title('Bubble Plot: GDP vs. Cost Index (Bubble Size = Unemployment Rate)')
plt.show()
```

- Countries with higher GDP per capita generally exhibit higher monthly incomes and purchasing power, but also tend to have higher cost of living indices.

- Unemployment rates vary widely, with only a few countries experiencing rates above 10%, while the majority maintain relatively low unemployment.
- Tourism plays a substantial role in the economies of certain countries, with some nations generating high receipts despite lower tourist numbers, highlighting the impact of spending per tourist.
- The pairplot and correlation heatmap demonstrate strong positive relationships between monthly income, purchasing power, and cost of living, suggesting that increases in one are often accompanied by increases in the others.
- The radar and bubble plots provide a comprehensive view of how countries perform across multiple economic dimensions, making it easier to identify outliers and leaders in specific areas.
- Overall, the dataset underscores the complexity of global economic dynamics and the importance of considering multiple indicators for a holistic understanding of a country's economic health.
